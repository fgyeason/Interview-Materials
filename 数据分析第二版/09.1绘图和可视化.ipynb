```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
fig = plt.figure()   
# ax1 = fig.add_subplot(2, 2, 1)
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
plt.plot(np.random.randn(50).cumsum(), 'k--')  #"k--"是一个线型选项，用于告诉matplotlib绘制黑色虚线图  放在最后一个没有被使用的subplot
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))  #图表的散点图方法
_ = ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
```

# 调整subplot周围的间距

```python
#wspace和hspace用于控制宽度和高度的百分比，可以用作subplot之间的间距。下面是一个简单的例子，其中我将间距收缩到了0
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```
